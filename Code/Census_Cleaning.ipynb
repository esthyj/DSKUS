```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('world_population.csv')
df.head()
```

```python
df = df.drop(['Country/Territory', 'Capital', 'Continent', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population'], axis=1)
df.head()
```

```python
df1 = pd.read_csv('world_pop.csv')
df1.head()
```

```python
merged_df = pd.merge(df1, df, on='CCA3')
merged_df.head()
```

```python
merged_df = merged_df.drop(['Series Name', 'Series Code', '2022 [YR2022]'], axis=1)
merged_df.head()
```

```python
merged_df = merged_df.rename(columns={'2018 [YR2018]': '2018 Population', '2019 [YR2019]': '2019 Population', '2020 [YR2020]': '2020 Population', '2021 [YR2021]': '2021 Population', 'Density (per km²)': 'Population Density 2022'})
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
new_column_order = ['Rank', 'CCA3', 'Country Name', '2018 Population', '2019 Population', '2020 Population', '2021 Population', '2022 Population', 'Area (km²)', 'Population Density 2022', 'Growth Rate', 'World Population Percentage']
merged_df = merged_df[new_column_order]
merged_df.head()
```

```python
tobacco_df = pd.read_csv('data/tobacco.csv')
tobacco_df.head()
```

```python
alcohol_df = pd.read_csv('data/alcohol.csv')
alcohol_df.head()
```

```python
alcohol_df = alcohol_df.drop(['Country'], axis=1)
alcohol_df.head()
```

```python
merged_df1 = pd.merge(tobacco_df, alcohol_df, on='CCA3')
merged_df1.head()
```

```python
merged_df1 = merged_df1.drop(['Country'], axis=1)
merged_df1.head()
```

```python
final_df = pd.merge(merged_df, merged_df1, on='CCA3')
final_df.head()
```

```python
final_df = final_df.replace('..', np.nan)
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['2018 Population'] = final_df['2018 Population'].astype(int)
final_df['2019 Population'] = final_df['2019 Population'].astype(int)
final_df['2020 Population'] = final_df['2020 Population'].astype(int)
final_df['2021 Population'] = final_df['2021 Population'].astype(int)
final_df['2018 Prevalence of current tobacco use (% of adults)'] = final_df['2018 Prevalence of current tobacco use (% of adults)'].astype(float)
final_df['2019 Prevalence of current tobacco use (% of adults)'] = final_df['2019 Prevalence of current tobacco use (% of adults)'].astype(float)
final_df['2020 Prevalence of current tobacco use (% of adults)'] = final_df['2020 Prevalence of current tobacco use (% of adults)'].astype(float)
final_df['2018 Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)'] = final_df['2018 Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)'].astype(float)
```

```python
final_df = final_df.drop(['2018 Prevalence of current tobacco use (% of adults)', '2019 Prevalence of current tobacco use (% of adults)', '2020 Prevalence of current tobacco use (% of adults)', '2018 Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)'], axis = 1)
final_df.head()
```

```python
final_df.to_csv('dataset.csv', index=False)
```
