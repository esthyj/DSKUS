```python
df_dist=pd.read_csv("C:/Users/jeony/OneDrive/바탕 화면/3학년 1학기/글로벌 인턴십/Group6/Data/dist_cepii/dist_cepii.csv")
df_dist
```

```python
df_cont=pd.read_csv("C:/Users/jeony/OneDrive/바탕 화면/3학년 1학기/글로벌 인턴십/Group6/Data/geo_cepii.csv")
df_cont
```

```python
df_dist1=df_dist.loc[:,['iso_o','iso_d','distcap']]
df_dist1
```

```python
df_cont1=df_cont.loc[:,['iso3','country']]
df_cont1
```

```python
df1=pd.merge(df_dist1, df_cont1, left_on='iso_o', right_on="iso3", how='inner')
df1
```

```python
df2=df1.rename(columns={'country':'country_o'})
df2
```

```python
df3=df2.loc[:,['distcap','country_o','iso_d']]
df3
```

```python
df4=pd.merge(df3, df_cont1, left_on='iso_d', right_on="iso3", how='inner')
df4
```

```python
df5=df4.loc[:,['country_o','country','distcap','iso_d']]
df5
```

```python
df6=df5.rename(columns={'country':'country_d'})
df6
```

```python
df6.isnull().sum()
```

```python
df7 = df6.drop_duplicates()
df7
```

```python
df8= df7[df7['country_o'] != df7['country_d']]
df8
```
